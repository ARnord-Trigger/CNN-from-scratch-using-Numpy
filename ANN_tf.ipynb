```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import RMSprop
import numpy as np
import pandas as pd
```

```python

# Load and preprocess the data
pf = "/content/RMBipolar02082023_081612.csv"
df = pd.read_csv(pf)

RM = df.values
RM = RM[:, 1:]

# Define the input and output matrices
X_train = np.identity(48)
Y_train = RM[0:48, 0:56].T

# Normalize the input data (optional)
X_train = (X_train - np.mean(X_train, axis=0)) / np.std(X_train, axis=0)

# Define the architecture of the ANN
def create_ann(input_dim, output_dim, hidden_layers):
    model = Sequential()
    model.add(Dense(hidden_layers[0], input_dim=input_dim, activation='relu'))

    for neurons in hidden_layers[1:]:
        model.add(Dense(neurons, activation='relu'))

    model.add(Dense(output_dim, activation='linear'))
    return model

# Parameters
input_dim = 48
output_dim = 56
hidden_layers = [108, 96]  # Customize the hidden layers as needed

# Create the model
model = create_ann(input_dim, output_dim, hidden_layers)

# Compile the model
model.compile(optimizer=RMSprop(learning_rate=0.001), loss='mse')

# Train the model
epochs = 200
batch_size = 12

model.fit(X_train, Y_train.T, epochs=epochs, batch_size=batch_size)

# Evaluate the model
mse_train = model.evaluate(X_train, Y_train.T)
print(f"Mean Square Error on Training Set: {mse_train:.4f}")

# Predict using the model
Y_pred = model.predict(X_train)

# Calculate R2 score
def r2_score(y_true, y_pred):
    y_true_mean = np.mean(y_true)
    ss_tot = np.sum((y_true - y_true_mean) ** 2)
    ss_res = np.sum((y_true - y_pred) ** 2)
    r2 = 1 - (ss_res / ss_tot)
    return r2

r2 = r2_score(Y_train, Y_pred.T)
print(f"R2 Score on Training Set: {r2:.4f}")
```
